```python
from sacla import *
```

```python
l=log('sacla2020.csv')
```

```python
import sys
sys.version_info
```

```python
import collections
runs=collections.defaultdict(list)
darks=[]
path="/UserData/gorkhover/TAIS2020/crystal"
for _,s in l[100:].iterrows():
    name=str(s['Name'][0])
    if not 'dark' in name: continue
    print(name,int(s['Run start']),"-",int(s['last run']))
    for i in range(int(s['Run start']),1+int(s['last run'])):
        try:
            t=h5py.File(f'{path}/{i}.h5','r')
            if not 'detector_2d_assembled_2' in list(t[list(t.keys())[1]].keys()): 
                print(f'{i}: missing ocatal')
        except Exception as e:
            print(f'{i}: error')
            print(e)
        else: 
            print(i)
            darks.append(i)
        
```

```python
torun=[[darks[i],darks[i+1]] for i in range(0,len(darks)-1)]
torun.append([darks[0]])
torun.append([darks[-1]])
torun
```

```python
torun=[[919786]]
```

```python
inputpath="/UserData/gorkhover/TAIS2020/crystal/{number}.h5"
outputpath="/UserData/gorkhover/TAIS2020/crystal/bg_{name}.h5"
cmds=[]

for r in torun:
    name='_'.join([str(e) for e in r])
    inputfiles=' '.join([inputpath.format(number=str(i)) for i in r])
    outputfile=outputpath.format(name=name)
    cmd=f"python ~/zimmf/create_dark.py {inputfiles} {outputfile}"
    cmds.append((name,cmd))
```

```python
for name,cmd in cmds:
    print(f"bg_{name}")
    print(cmd)
    print(qsub(['eval "$(conda shell.bash hook)"','conda activate zimmf',cmd],jobname=f"bg_{name}"))
    print()
```
